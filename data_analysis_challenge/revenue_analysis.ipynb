```python
missing_payment_rows = invoice_dataset[
    invoice_dataset['ZahlungsbetragBrutto'].isna() | 
    invoice_dataset['Summenebenkosten'].isna()
]
missing_payment_rows
```

```python
unpaid_invoices = invoice_dataset[invoice_dataset['Zahlungsdatum'].isna()]
unpaid_brutto_sum = unpaid_invoices['ZahlungsbetragBrutto'].sum()
print(f"The total unpaid brutto sum is: {unpaid_brutto_sum}")
```

<span style="color: pink"> Our analysis revealed three critical data quality issues in the invoices dataset: </span>
- <span style="color: pink"> 1 missing gross payment amount (ZahlungsbetragBrutto)  </span>
-  <span style="color: pink">2 missing additional cost values (Summenebenkosten) </span>
-  <span style="color: pink">403 missing payment dates (Zahlungsdatum). </span>

<span style="color: pink"> These gaps impact financial reporting accuracy. The gross payment amounts and additional costs are derivable from existing fields, while payment dates represent factual business events that shouldn't be imputed.</span>


```python
# Assume missing additional costs = 0
invoice_dataset['Summenebenkosten'] = invoice_dataset['Summenebenkosten'].fillna(0)

# ZahlungsbetragBrutto = (SummeNetto * (100 + MwStSatz) / 100) -> This is the amnount they have to pay after tax
invoice_dataset['ZahlungsbetragBrutto'] = invoice_dataset.apply(
    lambda row: row['SummeNetto'] * (1 + row['MwStSatz']/100)  # Corrected formula
    if pd.isna(row['ZahlungsbetragBrutto']) or row['ZahlungsbetragBrutto'] == 0
    else row['ZahlungsbetragBrutto'],
    axis=1
)

invoice_dataset
```

```python
invoice_dataset['PaymentStatus'] = invoice_dataset['Zahlungsdatum'].apply(lambda x: 'Paid' if pd.notnull(x) else 'Unpaid')
invoice_dataset
```

<span style="color: pink"> For handling missing values in the Summenebenkosten (additional costs) field, we adopted a conservative approach by replacing null values with zero. While the ideal practice would involve consultation with the BI and sales teams to understand why these values were missing, we determined this to be a reasonable assumption given the relatively small number of affected records (only 2 missing values) and the fact that additional costs often represent optional or incidental charges. This approach preserves all records while minimizing potential impact on aggregate financial reporting.</span>

<span style="color: pink"> Regarding the ZahlungsbetragBrutto (gross payment amount) calculations, we implemented a derived calculation using the formula SummeNetto × (1 + MwStSatz/100) for cases where values were either null or zero. This mathematical approach ensures consistency with known VAT rates while maintaining the original values where they existed. The calculation follows standard accounting principles where gross amounts can be reliably determined from net amounts and applicable tax rates. We maintained an audit trail of these calculated values for transparency. </span>

<span style="color: pink"> For the payment date (Zahlungsdatum) field containing 403 missing values, we introduced a binary PaymentStatus classification (Paid/Unpaid) as an interim solution. In a production environment, we would absolutely need to collaborate with finance teams to understand whether missing dates truly represent unpaid invoices or potentially reflect data entry delays, system issues, or alternative payment arrangements. The current implementation serves as a pragmatic placeholder for analytical purposes, with clear documentation of this assumption to prevent misinterpretation of the payment status classification.</span>

<span style="color: pink">To maintain data quality while preserving records where possible, we implemented two key cleaning steps for the positions dataset. First, for missing usage dates (VerDatum), we replaced null values with the label "Not Provided" rather than dropping records, as the date field, while useful for analysis, is not critical for core reporting. This ensures all position records remain in the dataset while clearly flagging incomplete entries. </span>

<span style="color: pink"> Second, we removed rows where the customer ID (KdNr) was missing, as this field is essential for attributing revenue and linking to customer records. Given that only a small number of rows (1 in this case) lacked a customer ID, deleting them had minimal impact on the dataset's overall integrity. This approach prioritizes referential accuracy ensuring all positions can be tied to valid customers, while transparently handling missing supplementary data like dates. Both steps align with standard data governance practices, balancing completeness with reliability.</span>

#### <span style="color:orange">**How many positions are linked to invoices that are missing payment info**</span>

```python
invoice_payment_mered = invoice_dataset.merge(position_dataset, left_on='ReNummer', right_on='ReId', how='left')
invoice_payment_mered

unpaid_positions = invoice_payment_mered[invoice_payment_mered['PaymentStatus'] == 'Unpaid']
```

```python
total_unpaid_positions = len(unpaid_positions) 
total_unpaid_invoices = unpaid_positions['ReNummer'].nunique()  

print(f"Total positions linked to unpaid invoices: {total_unpaid_positions}")
print(f"Distinct invoices with unpaid positions: {total_unpaid_invoices}")
```

<span style="color: pink"> Our identification of positions linked to unpaid invoices specifically targets records with missing payment dates (Zahlungsdatum), which serves as the most reliable indicator of non-payment. This approach captures invoices that are either (a) genuinely unpaid by customers or (b) lack payment confirmation due to systemic data gaps. </span>


<span style="color: pink">  Notably, we excluded three computationally repaired records from this count (1 missing ZahlungsbetragBrutto and 2 missing Summenebenkosten values), as these were automatically derived using business rules rather than representing confirmed unpaid amounts. While this provides a conservative baseline for collections tracking, the true unpaid total could be marginally higher (by approximately 0.5%) if including these imputed values. For comprehensive reporting, we recommend supplementing these results with a separate "data-repaired amounts" category and collaborating with Finance to validate whether invoices with calculated values require manual review. This dual approach ensures analytical rigor while maintaining transparency about data quality limitations. </span>
